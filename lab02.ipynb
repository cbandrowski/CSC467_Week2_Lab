```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits import mplot3d
from skimage import io as io_url
```

```python
frame = io_url.imread('https://raw.githubusercontent.com/cbandrowski/CSC467_Week2_Lab/6c774e7630fab8f8d4161a6a52e17e09879fd354/Lena.png')
frame = np.float32(frame)/255
plt.imshow(frame)
```

```python
def manual_convolution(image,kernel):
  #gets the image height and width
  height, width = image.shape
  #blank output for image dtype float32 ensures precision before converting to integers
  output = np.zeros((height,width), dtype=np.float32)
  #3x3 kernels need neighoring pixels adds a border with value of 0 makes sure calculations are correct
  padded_image = np.pad(image, pad_width=1, mode='constant', constant_values=0)

  #starts at 1 to skip padding
  for i in range(1,height+1): #rows
    for j in range(1,width+1): #columns
      region = padded_image[i-1:i+2,j-1:j+2] # getting 3x3 region
      output[i-1,j-1] = np.sum(region*kernel) #multiply and sum is stored
  return np.clip(output,0,255) #checks if pixels values between 0 and 255 grayscale


```
